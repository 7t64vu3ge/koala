# AI Orchestrator Playground

This notebook demonstrates how to interact with the AI Orchestrator programmatically.

```python
from orchestrator import plan_task

# Define your complex task
prompt = "I need to build a new landing page for my SAAS product. The UI should be modern and dark-themed, and the backend needs a user registration system."

# Call the orchestrator
plan = plan_task(prompt)

if plan:
    print(f"\nSummary: {plan.plan_summary}\n")
    print("Subtasks:")
    for task in plan.subtasks:
        print(f"[{task.id}] Agent: {task.assigned_agent.upper()} | {task.description}")
else:
    print("Failed to generate plan. Make sure your GROQ_API_KEY is configured in the .env file.")
```
